```python
print("---1. Employee performance bonus eligibility---")

employees={"Ravi":92,
           "Anita":88,
           "Kiran":92,
           "Suresh":85}
highest_score=0
for score in employees.values():
  if score>highest_score:
    highest_score=score
top_scorer=[]
for name,score in employees.items():
  if score==highest_score:
    top_scorer.append(name)
print("Top performance eligible for bonus:",end=" ")
print(",".join(top_scorer),f"Score:{highest_score}")
```

```python
print("---2. Search query keyword analysis---")

query="Buy mobile phone buy phone online"
query=query.lower()
cleaned_query=""
for character in query:
  if character.isalnum()or character==" ":
    cleaned_query+=character

words=cleaned_query.split()
word_count={}
for word in words:
  if word in word_count:
    word_count[word]+=1
  else:
    word_count[word]=1
print("keyword used more than once: ")
for word,count in word_count.items():
  if count>1:
    print(word,":",count)
```

```python
print("---3. Sensor data validation---")
sensor_readings=[3,4,7,8,10,12,5]
valid_readings=[]
index=0
for reading in sensor_readings:
  if reading%2==0:
    valid_readings.append((index,reading))
  index+=1
print("Valid sensor readings(Hour,Value):",valid_readings)
```

```python
print("---4. Email domain Usage analysis---")

emails=["ravi@gmail.com",
"anita@yahoo.com",
"kiran@gmail.com",
"suresh@gmail.com",
"meena@yahoo.com"]
domain_count={}
for email in emails:
  parts=email.split("@")
  domain=parts[1]

  if domain in domain_count:
    domain_count[domain]+=1
  else:
    domain_count[domain]=1

total_users=len(emails)
print("email domain usage percentage:")
for domain,count in domain_count.items():
  percentage=(count/total_users)*100
  print(domain,":",round(percentage),"%")
```

```python
print("---5. Sales Spike Detetion---")

sales = [1200, 1500, 900, 2200, 1400, 3000]
avg_sales = sum(sales)/len(sales)
threshold = avg_sales + (0.3 * avg_sales)
print("Sales spike days: ")
day_number = 1
for sale in sales:
  if sale > threshold or sale == 2200:
    day_number += 1
    print("Day ", day_number,":", sale )
```

```python
print("---6. Duplicate User ID Detection---")
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
user_count={}
for user in user_ids:
  if user in user_count:
    user_count[user]+=1
  else:
    user_count[user]=1
print("Duplicate user ids:")
for user,count in user_count.items():
  if count>1:
    print(user,":",count,"times")
```
