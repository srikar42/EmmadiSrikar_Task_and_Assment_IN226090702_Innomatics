```python
marks = [45, 78, 90, 33, 60]

pass_students_count = 0
fail_students_count = 0

for mark in marks:
    if mark >= 50:
        pass_students_count += 1
    else:
        fail_students_count += 1

print("Total Passed students:", pass_students_count)
print("Total Failed students:", fail_students_count)
```

```python
names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []

for name in names:
    cleaned_name = name.strip().lower()
    cleaned_names.append(cleaned_name)

print("Cleaned Names:", cleaned_names)
```

```python
messages = ["Hi", "Welcome to the platform", "OK"]

for message in messages:
    length = len(message)
    print(f"Message: '{message}' | Length: {length}")

    if length > 10:
        print("The Message is longer than 10 characters")
```
