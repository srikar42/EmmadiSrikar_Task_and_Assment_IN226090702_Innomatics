# Problem 2: Highest Salary from Employee Data

```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
highest_name = None
highest_salary = 0
for name,salary in dict.items(employees):
    if salary > highest_salary:
        highest_salary = salary
        highest_name = name
print("Highest Salary: ",highest_name,"-",highest_salary)
```

# Problem 3: Find Maximum and Minimum Values

```python
numbers = [45, 22, 89, 10, 66]
min = numbers[0]
max = numbers[0]
for i in numbers:
    if i < min:
        min = i
    if i > max:
        max = i
print(f"List:{numbers}\nmax:{max}\nmin:{min}")
```

# Problem 4: Count Products Above a Price Threshold

```python
prices = [450, 1200, 899, 1500, 300]
margin = 1000
count = 0
for i in prices:
    if i > margin:
        count+=1
print(f"Products above 1000: {count}")
```

# Problem 5: Calculate Attendance Percentage

```python
attendance = ["P", "P", "A", "P", "P"]
count = 0
for i in attendance:
    if i == "P":
        count+=1
attendance_percentage = count/len(attendance)*100
print(f"Attendance Percentage: {attendance_percentage}")
```

# Problem 10: Calculate Average Marks

```python
marks = [70, 75, 80, 85]

total = 0

for mark in marks:
    total += mark

average = total / len(marks)

print(f"Average Marks: {average}")
```
